```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#create a simple dataset
hours = np.array([2, 4, 6, 8, 10]).reshape(-1, 1)
marks = np.array([55, 60, 70, 80, 88])
hours, marks
```

```python
#Train a linear regrassion model
model = LinearRegression()
model.fit(hours, marks)
model
```

```python
#view learned parameter (y=mx+c)
print("m: ", model.coef_[0])
print("c: ", model.intercept_)
# marks = 4.3 * hours + 44.8
```

```python
#Visualize the best fit line
plt.scatter(hours, marks, color='blue', label = 'Actual Data')

predicated_marks = model.predict(hours)
plt.plot(hours, predicated_marks, color ='red', label='Best fit Line')

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study hours vs marks for linear regrassion")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# predict marks for new student
new_hours = np.array([[7]])
prediction = model.predict(new_hours)
print("Predicted marks: ", prediction[0])
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
```

```python
#Fit a Linear Regression Model (Closed-Form Solver)
lin_reg = LinearRegression()
lin_reg.fit(hours, marks)
print("m: ", lin_reg.coef_[0])
print("c: ", lin_reg.intercept_)
```

```python
# compute the cost(mean squared error) using library
predict_marks = lin_reg.predict(hours)
mse_lin = mean_squared_error(marks, predict_marks)
print("MSE/cost for linear regrassion: ", mse_lin)
```
